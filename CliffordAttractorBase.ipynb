```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros para el atractor de Clifford
a, b, c, d = -1, 7, -4, -3

# Número de puntos para el atractor
N = 200000

# Valores iniciales
x, y = 0, 0

# Listas para almacenar los puntos
x_vals = [x]
y_vals = [y]

# Generar el atractor
for i in range(N):
    x_next = np.sin(a * y) + c * np.cos(a * x)
    y_next = np.sin(b * x) + d * np.cos(b * y)
    x_vals.append(x_next)
    y_vals.append(y_next)
    x, y = x_next, y_next

# Graficar el atractor

plt.figure(figsize=(10, 10))
plt.scatter(x_vals, y_vals, s=0.5, c=x_vals, cmap='viridis', edgecolors='none')
plt.axis('off')
plt.title('Atractor de Clifford con Colores')
plt.colorbar(label='Valor de x')
plt.show()
```
